```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches
%matplotlib inline
```

## 1. Let's first get some data

```python
x = np.arange(-10,1,0.01)
```

```python
y = x/(1-x)
```

## 2. a simple plot

```python
fig1 = plt.figure(figsize=(7,6))

ax1 = plt.subplot(1,1,1)

ax1.plot(x, y, color='royalblue', linewidth=1.5)
ax1.plot(np.ones(20), np.arange(-10,10), linewidth=1.2, linestyle='-.', color='black')
ax1.plot(np.arange(-10,10), np.zeros(20)-1, linewidth=1.2, linestyle='-.', color='black')
ax1.plot(np.arange(0,0.501,0.1), np.ones(6), linewidth=1.2, linestyle='--', color='red')
ax1.plot(np.ones(11)*0.5, np.arange(0,1.01,0.1), linewidth=1.2, linestyle='--', color='red')
ax1.plot(np.arange(-1,0.01,0.1), np.ones(11)*-0.5, linewidth=1.2, linestyle='--', color='red')
ax1.plot(np.ones(6)*-1, np.arange(-0.5, 0.01,0.1), linewidth=1.2, linestyle='--', color='red')

ax1.arrow(-2.1, 0, 4, 0, head_width=0.1, head_length=0.2, facecolor='black')
ax1.arrow(0, -1.5, 0, 4.3, head_width=0.08, head_length=0.3, facecolor='black')

# This part helps you to remove the figure boundaries (or move these boundaries)
ax1.spines['left'].set_position(('data',0))
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.spines['bottom'].set_position(('data',0))
ax1.yaxis.set_ticks_position('left')


ax1.set_xlim([-2.1, 2.1])
ax1.set_ylim([-1.5, 3.1])
ax1.set_xticks([-2, -1, 0, 0.5, 1])
ax1.set_yticks([-1, -0.5, 1])

ax1.text(2.2, -0.2, r'$\frac{{\Delta}SWE}{P}$', va='center', size=20)
ax1.text(-0.2, 2.9, r'$\delta$', size=20)


plt.show()
plt.clf()
del(fig1)
```

## 3. now let's mark the different parts with differnt colors

```python
mask1 = (y<-0.5)
mask2 = (y>=-0.5)*(y<=0)
mask3 = (y>0)*(y<=1)
mask4 = (y>=0.999)
```

```python
fig1 = plt.figure(figsize=(7,6))

ax1 = plt.subplot(1,1,1)

ax1.plot(x[mask1], y[mask1], color='purple', linewidth=2)
ax1.plot(x[mask2], y[mask2], color='orange', linewidth=2)
ax1.plot(x[mask3], y[mask3], color='lightseagreen', linewidth=2)
ax1.plot(x[mask4], y[mask4], color='royalblue', linewidth=2)
ax1.plot(np.ones(20), np.arange(-10,10), linewidth=1.2, linestyle='-.', color='black')
ax1.plot(np.arange(-10,10), np.zeros(20)-1, linewidth=1.2, linestyle='-.', color='black')
ax1.plot(np.arange(0,0.501,0.1), np.ones(6), linewidth=1.2, linestyle='--', color='red')
ax1.plot(np.ones(11)*0.5, np.arange(0,1.01,0.1), linewidth=1.2, linestyle='--', color='red')
ax1.plot(np.arange(-1,0.01,0.1), np.ones(11)*-0.5, linewidth=1.2, linestyle='--', color='red')
ax1.plot(np.ones(6)*-1, np.arange(-0.5, 0.01,0.1), linewidth=1.2, linestyle='--', color='red')

ax1.arrow(-2.1, 0, 4, 0, head_width=0.1, head_length=0.2, facecolor='black')
ax1.arrow(0, -1.5, 0, 4.3, head_width=0.08, head_length=0.3, facecolor='black')


ax1.spines['left'].set_position(('data',0))
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.spines['bottom'].set_position(('data',0))
ax1.yaxis.set_ticks_position('left')


ax1.set_xlim([-2.1, 2.1])
ax1.set_ylim([-1.5, 3.1])
ax1.set_xticks([-2, -1, 0, 0.5, 1])
ax1.set_yticks([-1, -0.5, 1])

ax1.text(2.2, -0.2, r'$\frac{{\Delta}SWE}{P}$', va='center', size=20)
ax1.text(-0.2, 2.9, r'$\delta$', size=20)


#fig1.savefig('/raid1/chen423/serdp/data/paper3_analysis/plots/figS1.delta_function.png', dpi=600)
plt.show()
plt.clf()
```
